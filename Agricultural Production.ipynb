```python
def agricul():
    print('Area Planted/Desired:')
    area_planted = float(input())
    
    print('Cost of 50Kgs Seed:')
    standard_buy = float(input())
    
    print('Cost of 50Kgs Planting Fertiliser:')
    Fertiliser_Cost = float(input())
    
    print('Enter cost of Land and Menteinence:')
    Land = float(input())
    
    print('Enter cost of Labour:')
    labour = float(input())
    
    print('Enter Selling Price of 50Kgs :')
    standard_sell= float(input())
    
        #Constants
    standard_seed_weight = 10
    standard_seed_number = 120
    standard_area = 27
#     standard_sell = 2500
#     standard_buy = 3450
    standard_harvest = 180
    selling_weight = 50
    standard_fertiliser = 1.667
    FoliarFert = 13.5/1000
#     Fertiliser_Cost = 5450
    FoliarCost = 13.5
    
    #Chemicals
    #Standard_Area
    chem_area = 216
    #Standard_Chemical Solution
    chem_sol = 20
    
    #Insecticide
    pentagon_presc =10/1000
    pirimor_presc = 20/1000
    movento_presc = 10/1000
    
    #Fungicide
    fotress_presc = 50/1000
    milraz_presc = 50/1000
    ridomil_presc = 50/1000
    infinito_presc = 30/1000
    
    #Pre-Emergence Herbicide
    hurr_presc = 10/1000
    sencor_presc = 10/1000
    
    
    print('----Insecticide Pentagon-----')
    pentagon_price = 1426
    penatgon_tot = (area_planted/chem_area) * pentagon_presc
    pentagon_cost = (area_planted/chem_area) * pentagon_presc *   pentagon_price
    
    print('Pentagon Insecticide has been Applied',len(df[df['Insectcide']=='Pentagon']),'times')
    print('This is',(area_planted/chem_area )* chem_sol ,'Litres of Water-chemical solution')
    print('The total litres of Pentagon Insecticide used is', penatgon_tot*len(df[df['Insectcide']=='Pentagon']),'litres')
    print(penatgon_tot*len(df[df['Insectcide']=='Pentagon']),'Litres of Pentagon Cost',  pentagon_cost *len(df[df['Insectcide']=='Pentagon']),'Kshs')
    
    print('----Insecticide Pirimor-----')
    pirimor_price = 5080
    pirimor_tot = (area_planted/chem_area) * pirimor_presc
    pirimor_cost = (area_planted/chem_area) * pirimor_presc *   pirimor_price
    
    print('\nPirimor Insecticide has been Applied',len(df[df['Insectcide']=='Pirimor']),'times')
    print('This is',(area_planted/chem_area )* chem_sol ,'Litres of Water-chemical solution')
    print('The total litres of Pirimor Insecticide used is', pirimor_tot*len(df[df['Insectcide']=='Pirimor']),'litres')
    print(pirimor_tot*len(df[df['Insectcide']=='Pirimor']),'Litres of Pirimor Cost',  pirimor_cost *len(df[df['Insectcide']=='Pirimor']),'Kshs')
    
    print('----Insecticide Movento-----')
    movento_price = 5080
    movento_tot = (area_planted/chem_area) * movento_presc
    movento_cost = (area_planted/chem_area) * movento_presc *   movento_price
    
    print('\nMovento Insecticide has been Applied',len(df[df['Insectcide']=='movento']),'times')
    print('This is',(area_planted/chem_area )* chem_sol ,'Litres of Water-chemical solution')
    print('The total litres of Movento Insecticide used is', movento_tot*len(df[df['Insectcide']=='movento']),'litres')
    print(movento_tot*len(df[df['Insectcide']=='Pirimor']),'Litres of Movento Cost',  pirimor_cost *len(df[df['Insectcide']=='movento']),'Kshs')
    j=pentagon_cost *len(df[df['Insectcide']=='Pentagon'])+pirimor_cost *len(df[df['Insectcide']=='Pirimor'])+ pirimor_cost *len(df[df['Insectcide']=='movento'])
    print('\n----------------The total cost of Insecticide is',j,'KSh')
    
    
    
    print('----Fungicide Fotress Gold-----')
    fotress_price = 1272
    fotress_tot = (area_planted/chem_area) *  fotress_presc
    fotress_cost = (area_planted/chem_area) * fotress_presc *   fotress_price
    
    print('\nFotress Gold Fungicide has been Applied',len(df[df['Fungicide']=='Fotress Gold']),'times')
    print('This is',(area_planted/chem_area )* chem_sol ,'Kgs of Water-chemical solution')
    print('The total Kgs of Fotress Gold Fungicide used is',  fotress_tot*len(df[df['Fungicide']=='Fotress Gold']),'Kgs')
    print(fotress_tot*len(df[df['Fungicide']=='Fotress Gold']),'Kgs of Fotress Gold Cost',   fotress_cost *len(df[df['Fungicide']=='Fotress Gold']),'Kshs')
    
    print('----Fungicide Milraz-----')
    milraz_price = 5129
    milraz_tot = (area_planted/chem_area) *    milraz_presc
    milraz_cost = (area_planted/chem_area) *   milraz_presc *    milraz_price
    
    print('\nMilraz Fungicide has been Applied', len(df[df['Fungicide']=='Milraz']),'times')
    print('This is',(area_planted/chem_area )* chem_sol ,'Kgs of Water-chemical solution')
    print('The total Kgs of Milraz Fungicide used is',   milraz_tot*len(df[df['Fungicide']=='Milraz']),'Kgs')
    print( milraz_tot*len(df[df['Fungicide']=='Milraz']),'Kgs of Milraz Cost',milraz_cost *len(df[df['Fungicide']=='Milraz']),'Kshs')
    
    print('----Fungicide Ridomil Gold-----')
    ridomil_price = 200
    ridomil_tot = (area_planted/chem_area) *    ridomil_presc
    ridomil_cost = (area_planted/chem_area) *   ridomil_presc * ridomil_price
    
    print('\nRidomil Gold Fungicide has been Applied', len(df[df['Fungicide']=='Ridomil Gold']),'times')
    print('This is',(area_planted/chem_area )* chem_sol ,'Kgs of Water-chemical solution')
    print('The total Kgs of Ridomil Fungicide used is',    ridomil_tot *len(df[df['Fungicide']=='Milraz']),'Kgs')
    print( milraz_tot*len(df[df['Fungicide']=='Milraz']),'Kgs of Milraz Cost',  ridomil_cost*len(df[df['Fungicide']=='Ridomil Gold']),'Kshs')
    
    
    print('----Fungicide Infinito-----')
    infinito_price = 0
    infinito_tot = (area_planted/chem_area) *    infinito_presc
    infinito_cost = (area_planted/chem_area) *   infinito_presc * infinito_price
    
    print('\nInfinito Fungicide has been Applied', len(df[df['Fungicide']=='Infinito']),'times')
    print('This is',(area_planted/chem_area )* chem_sol *0,'Litres of Water-chemical solution')
    print('The total Kgs of Infinito Fungicide used is',    infinito_tot *len(df[df['Fungicide']=='Infinito']),'Litres')
    print(  infinito_tot *len(df[df['Fungicide']=='Infinito']),'Litres of Milraz Cost',infinito_cost*len(df[df['Fungicide']=='Infinito']),'Kshs')
    k=infinito_cost*len(df[df['Fungicide']=='Infinito']) +  ridomil_cost*len(df[df['Fungicide']=='Ridomil Gold']) +milraz_cost *len(df[df['Fungicide']=='Milraz']) +fotress_cost *len(df[df['Fungicide']=='Fotress Gold'])
                
    print('\n----------------The Total Cost of Fungicide is:',k,'-----')
    
    print('----Pre-Emergence Herbicide Hurricane-----')
    hurr_price = 1570
    hurr_tot = (area_planted/chem_area) *    hurr_presc
    hurr_cost = (area_planted/chem_area) *   hurr_presc * hurr_price
    
    print('\nHurricane Pre-Emergence Herbicide has been Applied', len(df[df['PreEmergenceChem']=='Hurricane']),'times')
    print('This is',(area_planted/chem_area )* chem_sol,'Litres of Water-chemical solution')
    print('The total Litres of Hurricane Herbicide used is',    hurr_tot *len(df[df['PreEmergenceChem']=='Hurricane']),'Litres')
    print( hurr_tot *len(df[df['PreEmergenceChem']=='Hurricane']),'Litres of Hurricane Cost', hurr_cost *len(df[df['PreEmergenceChem']=='Hurricane']),'Kshs')
    
    
    print('----Pre-Emergence Herbicide Sencor-----')
    sencor_price = 1570
    sencor_tot = (area_planted/chem_area) *    sencor_presc
    sencor_cost = (area_planted/chem_area) *   sencor_presc * sencor_price
    
    print('\nSencor Pre-Emergence Herbicide has been Applied', len(df[df['PreEmergenceChem1']=='Sencor']),'times')
    print('This is',(area_planted/chem_area )* chem_sol,'Litres of Water-chemical solution')
    print('The total Litres of Sencor Herbicide used is',sencor_tot *len(df[df['PreEmergenceChem1']=='Sencor']),'Litres')
    print( sencor_tot *len(df[df['PreEmergenceChem1']=='Sencor']),'Litres of Sencor Cost', sencor_cost *len(df[df['PreEmergenceChem1']=='Sencor']),'Kshs')
    
    l = hurr_cost *len(df[df['PreEmergenceChem']=='Hurricane'])+sencor_cost *len(df[df['PreEmergenceChem1']=='Sencor'])
    print('\n----------------The Total Cost of Pre-Emergence Herbicide is:',l,'Ksh')
    
    print('\n----TOTAL CHEMICAL COST-----',j+k+l,'---------')
                                           
    
    
    #Variable
   
#     Land = 10000
    chemicals = j+k+l
#     labour = 50000
    
    
    #Expression
    seed_weight =  (area_planted *standard_seed_weight) /(standard_area)
    seednumber = (area_planted  * standard_seed_number) /(standard_area)
    harvest = (area_planted *  standard_harvest)/(standard_area)
    cost = ( (seed_weight/selling_weight) *  standard_buy )/(1)
    fertil_weight = (area_planted * standard_fertiliser )/ standard_area
    Fertilizer_exp = ( fertil_weight/50) * Fertiliser_Cost
    Fertilizer=(Fertilizer_exp +(area_planted *   FoliarCost)/standard_area)
    return_weight_plant =   harvest/seednumber
    
  
     #Graphic1
    operating_profit = (harvest/selling_weight *  standard_sell) - (cost + Fertilizer + Land + chemicals + labour)
    labels = ['Revenue','Profit','Total Expense','Seed Cost','Fertiliser Expense','Land Lease','Chemicals Cost','Labour Cost']
    color = ['red','green','blue','orange','cyan']
    expn = cost + Fertilizer + Land + chemicals + labour
    Expense = [(harvest/selling_weight *  standard_sell),operating_profit,expn,cost, Fertilizer,Land,chemicals,labour]
    plt.figure(figsize=(15,4))
    plt.bar(height=Expense,x=labels,color=color)
#     plt.legend(loc='best')
    
#     Graphic 2
    print('------------------Expense------------------------------------------------------')
    plt.figure(figsize=(12,6))
    labelss = ['Seed Cost','Fertiliser Expense','Land Lease','Chemicals Cost','Labour Cost']
    color = ['red','green','blue','orange','cyan']
    expns = [cost,Fertilizer,Land,chemicals,labour]
    df_expns = pd.DataFrame(expns)
    df_expns.columns=['CostExp']

    df_lab = pd.DataFrame(labelss)
    df_lab.columns = ['Expense']
    df_lab_exp = pd.concat([df_lab ,df_expns],axis=1)
    
    df_lab_exp=df_lab_exp.set_index('Expense')
    print(df_lab_exp)
    df_lab_exp.plot.pie(y='CostExp', figsize=(5, 20),colors=color,title='Cost Expense Comparison',autopct='%.2f%%')
   
    expn = [cost,Fertilizer,Land ,chemicals,labour]
    color = ['red','green','blue','orange','cyan']
    
    #Output
    print('\n----------------------------1.SEED----------------------------------------------------------------------------')
    print('seed Weight required is',round(seed_weight,2),'Kgs that is',round(seed_weight/selling_weight,2),'Bags of 50Kgs seeds')
    print("The Number of Seeds(plant tubers) are",round(seednumber,0),"and the cost of seed is:",round(cost,2),'Kshs if 50 Kgs Cost:',
          standard_buy,'Kshs')
    print('\n---------------------------2.PLANTING FERTILIZER---------------------------------------------------------------------')
    print('\nPlanting Fertiliser Weight required is',round(fertil_weight,3),'Kgs this is',round(fertil_weight/50,2),'Bags of 50 Kgs' )
    print('\nThe Planting cost of Fertiliser is:',round(Fertilizer_exp,2),'Kshs',"50 Kgs Bag of Fertiliser is",
          Fertiliser_Cost,'Kshs if 50Kgs Bag Cost',Fertiliser_Cost)
    print('\n----------------------------3.FOLIAR--------------------------------------------------------------------')
    print('\nThe Amount in Litres of Foliar Fertiliser applied twice is',((area_planted *   FoliarFert)/ (standard_area)),'Litres')
    print('\nThe cost of Foliar Fertiliser applied twice in area',area_planted,'MetreSq is',
          ((area_planted *   FoliarCost)/standard_area)*2,'ksh if 1 litre Cost 1000 Kshs')
    print('------------------------5.HARVEST----------------------------------------------------------------------------')
    print('\nThe harvest of area',area_planted,'MetreSquare. is', round(harvest,2),'This is',harvest/selling_weight,'Bags',
          '\n and the income of these Bags:',round(harvest/selling_weight *  standard_sell,2),
          'Kshs if cost 50 Kgs Bags is:',standard_sell)
    print('\nReturn Weight Per Each Plant is', return_weight_plant,'Kgs')
    
    print('\n----------------------REVENUE-------------------------------------------------------------------------------')
    print('Revenue',(harvest/selling_weight *  standard_sell))
    print('\n-------------------------------OPERATION COST---------------------------------------------------------------')
    print('Total Expense',cost + Fertilizer + Land + chemicals + labour)
    print('-------------------------------OPERATING PROFIT---------------------------------------------------------------')
    print('Operating Profit of',area_planted,'MetreSq area is',round(operating_profit,2),'Ksh')
    print('\n---------------------PROFIT MARGIN--------------------------------------------------------------------------')
    print(((harvest/selling_weight *  standard_sell) - (cost + Fertilizer + Land + chemicals + labour))/
                       ((harvest/selling_weight *  standard_sell) ) *100)
    
    
```

```python
def fullprint(*args,**kwargs):
    from pprint import pprint
    import numpy
    opt = numpy.get_printoptions()
    numpy.set_printoptions(threshold = numpy.inf)
    pprint(*args,**kwargs)
    numpy.set_printoptions(**opt)
agricul()
```
