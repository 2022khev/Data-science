```python
from sklearn.datasets import make_classification
```

```python
X, Y = make_classification(n_samples = 200, n_classes=2, n_features=10, n_redundant=0, random_state=1)
```

```python
X.shape, Y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
X_train.shape, Y_train.shape
```

```python
X_test.shape, Y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
rf = RandomForestClassifier(max_features=5, n_estimators=100)
```

```python
rf.fit(X_train, Y_train)
```

```python
Y_pred = rf.predict(X_test)
Y_pred
```

```python
accuracy_score(Y_pred, Y_test)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
import numpy as np
```

```python
max_features_range = np.arange(1,6,1)
n_estimators_range = np.arange(10,210,10)
print(' max_features_range:\n',max_features_range)
print('\nn_estimators_range:\n',n_estimators_range)
```

```python
param_grid = dict(max_features=max_features_range, n_estimators=n_estimators_range)
```

```python
rf = RandomForestClassifier()
```

```python
grid = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5)
```

```python
grid.fit(X_train, Y_train)
```

```python
# print("Optimal parameters %s accuracy score of  %0.2f"
#       % (grid.best_params_, grid.best_score_))
grid.best_params_ ,grid.best_score_
```

```python
#Options # penalty='l2'   , tol=0.0001 , C=1.0 , max_iter=100


# Necessary imports
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

# Creating the hyperparameter grid
c_space = np.logspace(-5, 8, 15)
max_it = np.arange(10,160,10)
tol_n = [0.0001,0.005,0.01,0.5]

param_grid = {'C': c_space , 'max_iter': max_it, 'tol':tol_n }

# Instantiating logistic regression classifier
logreg = LogisticRegression()

# Instantiating the GridSearchCV object
logreg_cv = GridSearchCV(logreg, param_grid, cv = 5)

logreg_cv.fit(X_train, Y_train)

# Print the tuned parameters and score
print("Tuned Logistic Regression Parameters: {}".format(logreg_cv.best_params_))
print("Best score is {}".format(logreg_cv.best_score_))
```

```python
# Necessary imports
from scipy.stats import randint
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import RandomizedSearchCV
 
# Creating the hyperparameter grid
param_dist = {"max_depth": [3,None],
              "max_features": randint(1,9),
              "min_samples_leaf": randint(1, 9),
              "criterion": ["gini", "entropy"]}
 
# Instantiating Decision Tree classifier
tree = DecisionTreeClassifier()
 
# Instantiating RandomizedSearchCV object
tree_cv = RandomizedSearchCV(tree, param_dist, cv = 5,scoring='accuracy',return_train_score=True,verbose=10)
 
tree_cv.fit(X_train, Y_train)
 
# Print the tuned parameters and score
print("Tuned Decision Tree Parameters: {}".format(tree_cv.best_params_))
print("Best score is {}".format(tree_cv.best_score_))
```
