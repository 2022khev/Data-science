```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import seaborn as sns
```

# 1 Matplotlib   2seaborn  3pandas.plot  4plotly 5plot 6matplotlib axes

#  1. Matplotlib

```python
 #Matplotlib
colors = ['#1888','#2894','#6783']
weight = [35,78,45,89]
labels =['AB','WB','SB','IB']
wp = { 'linewidth' : 1, 'edgecolor' : "blue" }
explode = (0.1, 0.0, 0.1, 0.3)

plt.pie(weight,colors=colors,
        labels=labels,autopct = '%1.1f%%',
        pctdistance=0.5,startangle=90,
        textprops={'color':'grey','weight':'light','fontsize':'large'},
        wedgeprops = wp,explode = explode,
        shadow = False,
               )
# plt.legend(loc=4, title='Claim Types');
plt.legend(labels,title='Labels',
          loc ="center left",bbox_to_anchor =(1, 0, 0.5, 1))
plt.title('Comparison');
```

# 2. seaborn

```python
weight = [35,78,45,89]
labels =['AB','WB','SB','IB']
palette_color = sns.color_palette('dark')

plt.pie(weight,labels=labels ,colors=palette_color, autopct='%.1f%%',
#         textprops= {'color':'w','weight':'light','fontsize':'large'},
        explode=[0.2,0.01,0.01,0.01])

plt.legend(labels,title='Claim Types',
          loc ="center left",bbox_to_anchor =(0.0,0.0,0.5,0.0))
plt.title('Comparison');
```

# 3.pandas.plot

```python
import pandas as pd
weight = [35,78,45,89]
labels =['AB','WB','SB','IB']
explode = [0.01,0.01,0.01,0.3]
color = ['red','purple','indigo','violet']
av =pd.Series(weight,labels)
av.plot(kind='pie',
        autopct='%.1f%%',
        explode = explode,
        startangle=90,
        pctdistance=0.6,
        colors=color,
        textprops= {'color':'w'},
       title = 'Claim Proportion')
plt.legend(loc=3,title='claims');
```

# 4. plotly1

```python
import plotly.express as px

av =pd.Series(weight,labels)
av = pd.DataFrame(av,columns=['Weight'])
av.index.name = 'labels'
av.reset_index(inplace=True)
av

fig = px.pie(av,values='Weight',
             names = av['labels'],
             color= av['labels'],
              color_discrete_map = {'AB':'lightcyan',
                                     'SB':'cyan',
                                     'IB':'royalblue',
                                     'WB':'darkblue'
                                    },
             title='Comparison'
            )
fig.show(loc='left')
```

# 5. plotl2

```python
import plotly.graph_objects as go
fig = go.Figure(data = [go.Pie(
                labels =av['labels'].values,
                values= av['Weight'].values,             
                textinfo = 'label+percent',
                textfont_size=20,
                insidetextorientation='radial',
                 pull=[0.09, 0.01, 0.2, 0.01],
                title='Comparison')])
# option
# fig = go.Figure(data=[go.Pie(labels=labels, values=values, pull=[0, 0, 0.2, 0])])
# fig.show()
                
fig.show(loc='left')
```

# 6.matplotlib axes plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('Equal')
weight = [35,78,45,89]
labels =['AB','WB','SB','IB']
fig.suptitle('Comparison of Weight Proportion')
plt.xlabel('',fontsize=15,color='b',fontweight='bold')
fig.suptitle('Comparison of labels',fontsize=12,color='black',fontweight='bold')
ax.pie(weight,labels=labels,autopct = '%.2f%%',explode=[0.1,0.01,0.01,0.01]);
```
