```python
import requests
import bs4
from bs4 import BeautifulSoup
'https://www.elevolt.co.ke/jobs'
```

```python
url = "https://www.elevolt.co.ke/jobs"
```

```python
headers = {
    
    "User-Agent":"Mozilla/5.0 (windows NT 10.0;win64) AppleWebKit/537.36 (KHTML,like Gecko) Chrome/81.0.4044 Safari/537.36"
}
```

```python
r  = requests.get(url,{'headers':headers})
```

```python
import pandas as pd
soup = bs4.BeautifulSoup(r.text,'html.parser')
```

```python
soup.find_all()
```

```python
jobs = []
import numpy as np
k=np.arange(0,15)
for i in k:
    job=soup.find_all('h3')[i].text
    jobs.append(job)
jobs = pd.DataFrame(jobs,columns=['Jobs'])  
jobs
```

```python
organization = []
k=np.arange(2,15)
for i in k:
    org=soup.find_all('h5')[i].text
    organization.append(org)
Organization = pd.DataFrame(organization,columns=['Organization'])  
Organization
```

```python
pd.concat([jobs,Organization],axis=1)
```
